### Action item 1  (Filling the null values with median)

```python
df.describe().T
```

```python
 df.columns
```

```python
num_df =  df.select_dtypes(include ="number")
cat_df =  df.select_dtypes(exclude ="number")
```

```python
cat_df.columns
```

```python
num_df.isnull().sum()
```

```python
for column in num_df.columns:
    num_df[column]=num_df[column].fillna(num_df[column].median())
```

```python
num_df.isnull().sum()
```

```python
for column in cat_df:
    mode_value = (cat_df[column].mode().values[0])
    cat_df[column]= cat_df[column].fillna(mode_value)
                  
```

```python
cat_df.isnull().sum()
```

```python
df=pd.concat([num_df,cat_df],axis =1)
df.head(1)
```

### Action item 2(data type conversion)

```python
df['total_reviews'] = df['total_reviews'].astype(int)
df['purchased_last_month'] = df['purchased_last_month'].astype(int)
```

```python
df['delivery_date']=pd.to_datetime(df['delivery_date'],errors ='coerce')
df['data_collected_at']=pd.to_datetime(df['data_collected_at'],errors ='coerce')
```

### Action  item 3 (Dropping column with nulls percentage is higher)

```python
df.drop(['sustainability_tags'],axis =1,inplace = True)
```

### Feature Engineering 

```python
df['discounted_amount']=df['original_price'] - df ['discounted_price']
df.head(2)
```

### Observations: We created column called discounted amount that customer received with the help of original price and discounted price

```python
df['estimated_revenue']=df['discounted_price']*df['purchased_last_month']
df.head(2)
```

### Observations:We calculated estimated revenue that company received from sales from last month for each catogery of products


```python
df['percentage_range']=pd.cut(x=df['discount_percentage'],bins =[0,30,50,70,90],labels =["less","medium","high","veryhigh"])
df.head(2)
```
